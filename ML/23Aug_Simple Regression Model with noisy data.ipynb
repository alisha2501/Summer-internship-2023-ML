#### Objective: To build a linear model given by Ypred =mX+c

#### Generate noisy data points using Y=3X+4 with a noise having normal distribution with mean=0 and standard deviation 2

```python
import numpy as np
import matplotlib.pyplot as plt
X=np.linspace(0,10,100)
noise=np.random.normal(0,2,100)
plt.plot(noise)
plt.show()
```

```python
Y=3*X+4+noise
plt.plot(Y)
plt.show
```

```python
### Visualization of datapoints :
plt.plot(X,Y)
plt.show()
```

```python
plt.plot(X,Y,'*')
plt.xlabel('Input data points')
plt.ylabel('Output data points')
plt.title('Datapoints')
plt.show()
```

```python
X1=[1,2,3]
Y1=[1,2,3]
```

```python
plt.plot(X1,Y1)
plt.show()
```

```python
plt.scatter(X1,Y1)
plt.show()
```

#### We need to split the data points into training points and testing points

```python
from sklearn.model_selection  import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.3)
```

```python
xtest.ndim
```

```python
xtest.shape
```

```python
xtest.shape
```

```python
xtrain.shape
```

### Build a Linear Regression model

```python
from sklearn.linear_model import LinearRegression
```

```python
LR=LinearRegression()
```

```python
xtrain.shape
```

#### convert the given dataset to 2D set so that the fit function can be used.

```python
xtrain=xtrain.reshape(-1,1)
ytrain=ytrain.reshape(-1,1)
```

```python
xtrain.shape
```

```python
LR.fit(xtrain,ytrain)
```

#### Testing phase: Validate the model using predict function from Linearregression class 

```python
xtest=xtest.reshape(-1,1)
ypred=LR.predict(xtest)
```

#### Visualize using pyplot functionality by plotting (ypredict and ytest versus xtest)

```python
plt.scatter(xtest,ypred)
plt.scatter(xtest,ytest)
plt.xlabel("Test input data")
plt.ylabel("Test and predict output data")
plt.title("Testing Phase")
plt.show()
```

```python
plt.scatter(xtest,ypred,color='red',label='Ypred')
plt.scatter(xtest,ytest,color='blue',label='Ytest')
plt.xlabel("Test input data")
plt.ylabel("Test and predict output data")
plt.title("Testing Phase")
plt.legend()
plt.show()
```

#### To know about the Parameters of the Linear Model

```python
LR.coef_
```

```python
LR.intercept_
```

#### Estimate the output of the model for the given input X=30.2

```python
x1=30.2
## convert the given value to an numpy array
x1=np.array(x1)
## then convert 1D data to 2D data
x1=x1.reshape(-1,1)
LR.predict(x1)
```

```python
ypred.shape
```

#### Performance matrix for the linear model.

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,mean_absolute_percentage_error,d2_absolute_error_score
mean_absolute_error(ytest,ypred)
mean_squared_error(ytest,ypred)
mean_absolute_percentage_error(ytest,ypred)
d2_absolute_error_score(ytest,ypred)
```

```python
print(f"Mean Absolute error of the linear regression model : {mean_absolute_error(ytest,ypred)}")
print(f"mean_squared_errore error of the linear regression model : {mean_squared_error(ytest,ypred)}")
print(f"Mean Absolute percentage error of the linear regression model : {mean_absolute_percentage_error(ytest,ypred)}")
print(f"d2 absolute error of the linear regression model : {d2_absolute_error_score(ytest,ypred)}")
```
